```python
import pandas as pd
import numpy as np
```

```python
db_url = "postgresql://robot-startml-ro:pheiph0hahj1Vaif@postgres.lab.karpov.courses:6432/startml"

users_df = pd.read_sql("SELECT * FROM public.user_data", db_url)
posts_df = pd.read_sql("SELECT * FROM public.post_text_df", db_url)
feed_df = pd.read_sql("SELECT * FROM public.feed_data LIMIT 10000", db_url)
```

```python
# Merge all tables

df = pd.merge(
    feed_df,
    users_df,
    on='user_id',
    how='left'
)

df = pd.merge(df,
             posts_df,
             on='post_id',
             how='left')

df.sort_values('timestamp', inplace=True)
df.head()
```

```python
# Add new feature

df['post_len'] = df['text'].apply(lambda x: len(x))
```

```python
object_cols = ['action', 'country', 'city', 'os', 'source', 'topic']

for col in object_cols:
        one_hot = pd.get_dummies(df[col], prefix=col, drop_first=True)     
        df = pd.concat((df.drop(col, axis=1), one_hot), axis=1)
```

```python
train_new = df.iloc[:-2000].copy()
test_new = df.iloc[-2000:].copy()

X_train = train_new.drop(['target', 'timestamp', 'user_id', 'post_id', 'text'], axis=1)
X_test = test_new.drop(['target', 'timestamp', 'user_id', 'post_id', 'text'], axis=1)

y_train = train_new['target']
y_test = test_new['target']
```

```python
from catboost import CatBoostClassifier, Pool

catboost = CatBoostClassifier()

catboost.fit(X_train, y_train)
```
